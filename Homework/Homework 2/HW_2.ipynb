## 1. Matrix Rank

For parts (a) and (b), we will assume that we have a matrix $A \in \mathbb{R}^{m \times n}$ such that the row rank is equal to $r$ and a column rank equal to $k$.

### (a)

We assume that $A$ has $k$ linearly independent columns, thus $A$ can be written as
$$A = CW = \begin{bmatrix} c_1 & ... & c_k\end{bmatrix}W$$

where $A_i = C * W_i$, $C \in \mathbb{R}^{m \times k}$ and $W \in \mathbb{R}^{k \times n}$. We define $c$ such that it spans the range of $A$. We can write $c$ in row form such that

$$A = \begin{bmatrix} \bar{c}_1^T \\ ... \\ \bar{c}_m^T \end{bmatrix}W$$
$$A = \begin{bmatrix} \bar{c}_1^T W \\ ... \\ \bar{c}_m^T W \end{bmatrix}$$

We can see that each row in the resulting $A$ is a linear combination of the matrix $W$. Because every row is a linear combination of $W$ and there are only $k$ rows in $W$, we can see that the row rank of $A$, $r$, must be less than or equal to $k$ - i.e. $r \leq k$.

### (b)

We assume that $A$ has $r$ linearly independent rows, thus $A$ can be written as
$$A = VR = \begin{bmatrix} v_1 \\ ... \\ v_r \end{bmatrix} R$$

where $\bar{A}_i^T = v_i^T R$, $v \in \mathbb{R}^{m \times r}$ and $R \in \mathbb{R}^{r \times n}$. We define $R$ such that it is a basis for $A$. We can write $v$ in column form such that

$$A = \begin{bmatrix} \bar{v}_1^T & ... & \bar{v}_m^T \end{bmatrix} R$$
$$A = \begin{bmatrix} \bar{v}_1^TR & ... & \bar{v}_m^TR \end{bmatrix}$$

We can see that each row in the resulting $A$ is a linear combination of the matrix $R$. Because every column is a linear combination of $R$ and there are only $r$ columns in $R$, we can see that the col rank of $A$, $k$, must be less than or equal to $r$ - i.e. $k \geq r$.

### (a)

Given $A = \begin{bmatrix} 1 & 2 \\ 3 & 4\end{bmatrix}$, $A^T = \begin{bmatrix} 1 & 3 \\ 2 & 4\end{bmatrix}$.

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].plot([0, 1], [0, 3], color='b', label='$\mathcal{R}(A^T)$', linewidth=2)
ax[0].plot([0, 2], [0, 4], color='b', linewidth=2)
ax[0].plot([0], [0], color='r', label='$\mathcal{N}(A)$', marker='.', markersize=10)

ax[0].set_xlim([-4, 4])
ax[0].set_ylim([-6, 6])
ax[0].grid()
ax[0].legend()
ax[0].set_title('Domain of $A$')

rect = patches.Rectangle([0, 0], 2, 4, fill=False, edgecolor='dimgrey', linewidth = 2)
rect_1 = patches.Rectangle([0, 0], 1, 3, fill=False, edgecolor='dimgrey', linewidth = 2)

ax[0].add_patch(rect)
ax[0].add_patch(rect_1)


ax[1].plot([0, 1], [0, 2], color='b', label='$\mathcal{R}(A)$', linewidth=2)
ax[1].plot([0, 3], [0, 4], color='b', linewidth=2)
ax[1].plot([0], [0], color='r', label='$\mathcal{N}(A^T)$', marker='.', markersize=10)

ax[1].set_xlim([-4, 4])
ax[1].set_ylim([-6, 6])
ax[1].grid()
ax[1].legend()
ax[1].set_title('Co-Domain of $A$')

rect = patches.Rectangle([0, 0], 1, 2, fill=False, edgecolor='dimgrey', linewidth = 2)
rect_1 = patches.Rectangle([0, 0], 3, 4, fill=False, edgecolor='dimgrey', linewidth = 2)

ax[1].add_patch(rect)
ax[1].add_patch(rect_1)

fig.tight_layout()

plt.show()
```
